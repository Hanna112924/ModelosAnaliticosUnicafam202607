# LISTAS

```python
# Mi lista
lista_número=[]
# Lista con varios elementos
lista_número=[1,2,3,4,5]
print(lista_número)
```

```python
# Agregar elementos a mi lista
lista_número.append("6")
print(lista_número)
```

```python
# Cambio
print("Actualiza")
valor_buscar=input("Por favor, ingrese el número para actualizar")
valor_reemplazar=input("Ingrese el valor a reemplazar")
lista_número[lista_número.index(valor_buscar)]=valor_reemplazar
print("Cambio ejecutado. Imprimiendo resultado")
print(lista_número)
```
