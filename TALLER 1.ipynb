# OPERADORES MATEMATICOS

```python
a=10
b=3
```

```python
# SUMA
a=10
b=3
c=a+b
print(a+b)
print(c)
var="Suma Total"
print(var)
```

```python
# RESTA
c=a-b
print(a-b)
print(c)
var="Resta total"
print(var)
```

```python
#  MULTIPLICACIÓN
c=a*b
print(a*b)
print(c)
var="Resultado"
print(var)
```

```python
# DIVISIÓN
c=a/b
print(a/b)
print(c)
var="Total"
print(var)
```

```python
# MODELO
print(a%b)
var="Total"
print(var)
```

```python
# POTENCIA
print(a**10)
var="Resultado final"
print(var)
```

```python
# DIVISIÓN ENTERA
print(a//b)
```

# TEXTO

```python
saludo="Hola, Mundo!"
print(saludo)
```

```python
# EXTRAER EL PRIMER CARACTER
print(saludo[0])
```

```python
# EXTRAER LOS ULTIMOS CARACTERES
print(saludo[6:12])
```

```python
# SALUDOS
saludo_completo="¿Cómo estas?"
print(saludo_completo)
```

```python
# CONCATENAR SALUDOS
palabra1="Hola, Mundo!"
palabra2="¿Cómo estas?"
print(palabra1+"."+palabra2)
```

```python
# COMBINAR FORMATO
print(saludo_completo.upper())
```

```python
# NUEVA VARIABLE SALUDO
saludo_mayuscula="¿CÓMO ESTAS?"
print(saludo_mayuscula)
```
