```python
# ¿Qué rama se ejecuta?

puntaje = 72
if puntaje >= 90:
  print("Sobresaliente")
elif puntaje >=80:
  print("Notable")
elif puntaje >=70:
  print("Aceptable")
elif puntaje >=60:
  print("Suficiente")
else:
  print("Reaprobado")
```

```python
# Clasificador de IMC

peso = 70
altura = 1.75
imc = peso/altura ** 2

print(f"IMC: {imc:.1f}")

if imc < 18.5:
  print("Bajo peso")
elif imc < 25:
  print("Normal")
elif imc < 30:
  print("Sobrepeso")
else:
  print("Obesidad")
```

```python
# Condicional con bugs

edad = 20
if edad == 18:
  print("Exactamente 18")
elif edad > 18:
  print("Mayor de 18")
else:
  print("Menor de 18")
```

```python
# Acceso a sistema

edad = 20
password = "abc123"

mayor = edad > 18
pass_ok =len(password) > 8

if mayor and pass_ok:
  print("Acceso permitido")
elif not mayor:
  print("Acceso denegdo: edad insuficiente")
elif not pass_ok:
  print("Acceso denegado: contraseña muy corta")

```

```python
# Calculadora de descuento

monto = 120000

if monto < 50000:
  descuento = 0
elif monto < 100000:
  descuento = 0.10
elif monto < 200000:
   descuento = 0.20
else:
  descuento: 0.30

ahorro = monto * descuento
total = monto - ahorro
print(f"Descuento: {int(descuento*100)}% (${ahorro:,.0f})")
print(f"total: ${total:,.0f}")
```

```python
# FizzBuzz clásico

def fizzbuzz(n):
  if n % 3 == 0 and n % 5 == 0:
    print("fizzBuzz")
  elif n % 3 == 0:
    print("Fizz")
  elif n % 5 == 0:
    print("Buzz")
  else:
    print(n)
```

```python
# Recorrer una lista

precios = [12000, 45000, 8500, 67000, 3200]
caros= 0

for i, p in enumerate(precios, 1):
  print(f"Item{i}: ${p:,}")
  if p > 10000:
    caros += 1

print(f"Articulos >$10,000: {caros}")
```

```python
# While con condición

saldo = 1000
mes = 0

while saldo > 0:
  saldo -=150
  saldo -=20
  mes += 1
  print(f"Mes {mes}: saldo ${saldo}")
  print(f"saldo agotado en {mes} meses")
```

```python
# Break y continue

numeros = [3, 7, -2, 15, 4, -8, 9, 1]

for n in numeros:
  if n < 0:
    continue
  if n > 10:
    print(f"Encontre {n}, deteniendo.")
    break
  print(f"Procesando: {n}")
```
