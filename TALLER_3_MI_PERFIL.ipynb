# DICCIONARIO

```python
mi_perfil={
"Nombre":"Ana",
"Apellido":"Nivia",
"Edad":"30",
"Ciudad":"Bogotá",
"Fecha de nacimiento":"04 dic 1995",
"Pasatiempo":["Leer","Entrenar","Ver peliculas"]}
```

```python
# Añadir un elemento nuevo
mi_perfil["Empresa"]="Amore Group SAS"
print(mi_perfil)
```

```python
# Actulización de datos
print("ACTUALIZACIÓN DE DATOS")
valor_buscar=input("Ingrese el apellido para actualizar:")
valor_reemplazar=input("Ingrese el nuevo valor:")
for clave, valor in mi_perfil.items():
  if valor == valor_buscar:
    mi_perfil[clave]=valor_reemplazar
    print("Cambio ejecutado")
    break
else:
  print("El valor no se encontro.")
  print(mi_perfil)
```

```python
# Actualizar un dato existente
mi_perfil["Apellido"]="García"
print(mi_perfil)
```

```python
# Eliminar un dato con del
del mi_perfil["Fecha de nacimiento"]
print(mi_perfil)
```

```python
# Eliminar un dato con pop
clave_pop=input("Ingresa la clave que deseas eliminar(usando.pop()):")
if clave_pop in mi_perfil:
  valor_eliminado = mi_perfil.pop(clave_pop)
  print(f"se ha eliminado'{clave_pop}'cuyo valor era:{valor_eliminado}" )
else:
  print ("error:la clave no existe.")
```

```python
# Consultar
clave_cons=input("Consultar un dato:")
resultado=mi_perfil.get(clave_cons,"No disponible")
print(f"Resultado:{resultado}")
```
